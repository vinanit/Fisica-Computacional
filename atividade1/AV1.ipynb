```python
#Solicitar o RA
#Calcular a soma dos digitos do RA
#Definir a soma dos digitos do RA
#Calcule e imprima a altura máxima atingida pelo projetil usando a expressão: (hmáx = V0**2 / 2g)
#Imprimir o valor 
#Solicitar o RA
Registro_aluno = str(input('Digite o seu RA: '))

#Calcular a soma dos digitos do RA
Soma_RA = 0
for i in range(len(Registro_aluno)):
    aux = int(str(Registro_aluno[i]))
    Soma_RA += aux

#Informar as variáveis e Calcular a altura maxima
velocidade_inicial = Soma_RA*2 #M/s
gravidade = 9.8 #M/s**2

altura_maxima = ((velocidade_inicial**2)/(2*gravidade))

#Imprimir o valor obtido
print(altura_maxima)
```
